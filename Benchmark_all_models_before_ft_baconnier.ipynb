## **BENCHMARK - Dataset Bercy (Baconnier) - 13 502 lignes - 10% du dataset complet**

```python
import os
import sys
import gc
import time
import random
from dataclasses import dataclass
from typing import Dict, List, Optional

import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

os.environ["WANDB_DISABLED"] = "true"
os.environ["TOKENIZERS_PARALLELISM"] = "false"
os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"

def install(package: str):
    os.system(f"{sys.executable} -m pip install -q {package}")

try:
    import datasets
    from sentence_transformers import SentenceTransformer
    from transformers import AutoModel, AutoTokenizer, AutoConfig
except ImportError:
    print("Installation des dépendances...")
    install("sentence-transformers datasets pandas matplotlib seaborn scikit-learn transformers")
    import datasets
    from sentence_transformers import SentenceTransformer
    from transformers import AutoModel, AutoTokenizer, AutoConfig

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Mode : {'GPU' if DEVICE == 'cuda' else 'CPU'}")

def set_all_seeds(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_all_seeds(42)

def cuda_empty_cache():
    if DEVICE == "cuda":
        torch.cuda.empty_cache()

# CONFIG
@dataclass
class Config:
    max_samples: int = None
    batch_size: int = 32
    max_length: int = 512
    block_size_cuda: int = 128
    block_size_cpu: int = 32
    k_eval: int = 20
    mrr_k: int = 10
    ndcg_k: int = 10

CFG = Config()

# 1) WRAPPER SNOWFLAKE (Transformers natif)
class SnowflakeWrapper:
    """
    Wrapper pour utiliser Snowflake (HF transformers) avec une API encode() proche SentenceTransformer.
    Utilise mean pooling + normalisation L2.
    """
    def __init__(self, model_name: str):
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        config = AutoConfig.from_pretrained(model_name, trust_remote_code=True)

        # Fix anti-crash potentiels (si présents)
        if hasattr(config, "use_memory_efficient_attention"):
            config.use_memory_efficient_attention = False
        if hasattr(config, "unpad_inputs"):
            config.unpad_inputs = False

        self.model = AutoModel.from_pretrained(
            model_name,
            config=config,
            add_pooling_layer=False,
            trust_remote_code=True
        ).to(DEVICE)
        self.model.eval()

    @staticmethod
    def _mean_pooling(last_hidden_state: torch.Tensor, attention_mask: torch.Tensor) -> torch.Tensor:
        mask = attention_mask.unsqueeze(-1).type_as(last_hidden_state)
        summed = (last_hidden_state * mask).sum(dim=1)
        denom = mask.sum(dim=1).clamp(min=1e-9)
        return summed / denom

    def encode(
        self,
        sentences: List[str],
        prompt_name: Optional[str] = None,
        batch_size: int = 32,
        show_progress_bar: bool = False,
        prefix_query: str = "query: ",
        prefix_doc: str = ""
    ) -> torch.Tensor:
        prefix = prefix_query if prompt_name == "query" else prefix_doc
        inputs = [prefix + s for s in sentences]

        all_embeddings = []
        for i in range(0, len(inputs), batch_size):
            batch_texts = inputs[i:i + batch_size]
            batch_tokens = self.tokenizer(
                batch_texts,
                padding=True,
                truncation=True,
                return_tensors="pt",
                max_length=CFG.max_length
            ).to(DEVICE)

            with torch.inference_mode():
                outputs = self.model(**batch_tokens)
                emb = self._mean_pooling(outputs[0], batch_tokens["attention_mask"])
                emb = torch.nn.functional.normalize(emb, p=2, dim=1)
                all_embeddings.append(emb)

        if all_embeddings:
            return torch.cat(all_embeddings, dim=0)
        return torch.empty((0, 0), device=DEVICE)

# 2) DATASET (MODIF: 10% COHÉRENT = 10% des termes ET 10% des defs)
print("\nChargement du dataset...")
try:
    dataset = datasets.load_dataset("baconnier/administration_multi-term-dataset", split="train")

    cols = dataset.column_names
    col_terme = next(c for c in cols if "term" in c.lower())
    col_def = next(c for c in cols if "def" in c.lower())

    # 1 TERME -> 1 DEF (définition canonique = la plus fréquente)
    from collections import Counter, defaultdict as dd

    TERM_TO_DEF_COUNTER = dd(Counter)

    for row in dataset:
        terms = row[col_terme]
        defs = row[col_def]

        m = min(len(terms), len(defs))
        for i in range(m):
            t = str(terms[i]).strip().lower()
            d = str(defs[i]).strip()
            if not t or not d:
                continue
            TERM_TO_DEF_COUNTER[t][d] += 1

    TERM_TO_DEF = {}
    for t, counter in TERM_TO_DEF_COUNTER.items():
        best_def, _ = counter.most_common(1)[0]
        TERM_TO_DEF[t] = best_def

    # liste complète des termes uniques (stable)
    TERMS_ALL = sorted(TERM_TO_DEF.keys())

    # 10% des termes (et donc 10% des defs)
    nq = len(TERMS_ALL)
    nq_10 = max(1, int(0.10 * nq))
    rng = np.random.default_rng(42)
    sel = rng.choice(nq, size=nq_10, replace=False)

    TERMS = [TERMS_ALL[i] for i in sel]

    # Documents = defs uniquement des termes sélectionnés
    DOCUMENTS = [TERM_TO_DEF[t] for t in TERMS]

    # Mapping : terme -> index (unique) dans DOCUMENTS
    TERM_TO_INDICES = {t: [i] for i, t in enumerate(TERMS)}

    # Questions alignées exactement sur TERMS
    QUESTIONS = [f"C'est quoi {t} ?" for t in TERMS]
    DATASET_QA = [(q, "") for q in QUESTIONS]

    print(f"Nb termes uniques total : {len(TERMS_ALL)}")
    print(f"Nb termes sélectionnés (10%) : {len(TERMS)}")
    print(f"Nb questions : {len(QUESTIONS)}")
    print(f"Nb documents (defs) : {len(DOCUMENTS)}")

    # DEBUG
    t0 = TERMS[0]
    print("Exemple terme :", t0)
    print("Index def :", TERM_TO_INDICES[t0])
    print("Extrait def :", DOCUMENTS[TERM_TO_INDICES[t0][0]][:160], "...")

except Exception as e:
    print(f"Erreur Dataset : {e}")
    DATASET_QA = []

# 3) CANDIDATS + PREFIXES
CANDIDATS = [
    {"name": "Solon-Large", "id": "OrdalieTech/SOLON-embeddings-large-0.1"},
    {"name": "Snowflake-Arctic", "id": "Snowflake/snowflake-arctic-embed-m-v2.0"},
    {"name": "BGE-M3", "id": "BAAI/bge-m3"},
    {"name": "E5-Large", "id": "intfloat/multilingual-e5-large"},
    {"name": "E5-Large-instruct", "id": "intfloat/multilingual-e5-large-instruct"},
    {"name": "GTE-Multilingual", "id": "Alibaba-NLP/gte-multilingual-base", "trust": True},
    {"name": "E5-Small", "id": "intfloat/multilingual-e5-small"},
]

MODEL_PREFIXES: Dict[str, Dict[str, str]] = {
    "E5": {"q": "query: ", "d": "passage: "},
    "Snowflake": {"q": "query: ", "d": ""},
}

# 4) METRICS + TOPK BLOCKWISE
def compute_retrieval_metrics_from_topk(
    topk_indices: torch.Tensor,
    questions: List[str],
    mrr_k: int = 10,
    ndcg_k: int = 10
) -> Dict[str, float]:
    n, K = topk_indices.shape
    r1 = r5 = r10 = r20 = 0
    mrr = 0.0
    ndcg = 0.0

    k5 = min(5, K)
    k10 = min(10, K)
    k20 = min(20, K)
    mrr_k = min(mrr_k, K)
    ndcg_k = min(ndcg_k, K)

    for i in range(n):
        term = (
            questions[i]
            .replace("C'est quoi", "")
            .replace("?", "")
            .strip()
            .lower()
        )

        relevant = set(TERM_TO_INDICES.get(term, []))
        ranked = topk_indices[i].tolist()

        if ranked[0] in relevant:
            r1 += 1
        if any(idx in relevant for idx in ranked[:k5]):
            r5 += 1
        if any(idx in relevant for idx in ranked[:k10]):
            r10 += 1
        if any(idx in relevant for idx in ranked[:k20]):
            r20 += 1

        rr = 0.0
        for rank_pos, doc_idx in enumerate(ranked[:mrr_k], start=1):
            if doc_idx in relevant:
                rr = 1.0 / rank_pos
                break
        mrr += rr

        dcg = 0.0
        for rank_pos, doc_idx in enumerate(ranked[:ndcg_k], start=1):
            if doc_idx in relevant:
                dcg += 1.0 / np.log2(rank_pos + 1)

        rel_count = min(len(relevant), ndcg_k)
        idcg = 0.0
        for rank_pos in range(1, rel_count + 1):
            idcg += 1.0 / np.log2(rank_pos + 1)

        ndcg += (dcg / idcg) if idcg > 0 else 0.0

    return {
        "R@1 (%)": (r1 / n) * 100,
        "R@5 (%)": (r5 / n) * 100,
        "R@10 (%)": (r10 / n) * 100,
        "R@20 (%)": (r20 / n) * 100,
        f"MRR@{mrr_k}": (mrr / n),
        f"nDCG@{ndcg_k}": (ndcg / n),
    }

def compute_topk_blockwise(
    emb_q: torch.Tensor,
    emb_d: torch.Tensor,
    k_eval: int,
    block_size: int
) -> torch.Tensor:
    nq = emb_q.size(0)
    nd = emb_d.size(0)
    K = min(k_eval, nd)
    emb_d_t = emb_d.T

    out_topk = []
    for i in range(0, nq, block_size):
        q_block = emb_q[i:i + block_size]
        scores = torch.matmul(q_block, emb_d_t)
        topk = torch.topk(scores, k=K, dim=1).indices
        out_topk.append(topk.detach().cpu())

        del scores, topk, q_block
        cuda_empty_cache()

    return torch.cat(out_topk, dim=0)

# 5) LOADER + BENCHMARK (SANS CACHE)
def load_model_smart(cand):
    t0 = time.perf_counter()
    model = None
    err = None

    try:
        if "Snowflake" in cand["name"]:
            model = SnowflakeWrapper(cand["id"])
        else:
            kw = {"torch_dtype": torch.float16} if DEVICE == "cuda" else {}
            trust = cand.get("trust", False)
            model = SentenceTransformer(
                cand["id"],
                trust_remote_code=trust,
                device=DEVICE,
                model_kwargs=kw
            )
    except Exception as e:
        err = e

    t1 = time.perf_counter()
    load_time = t1 - t0
    return model, load_time, err

def run_benchmark(model, model_name: str) -> Dict[str, float]:
    questions = QUESTIONS
    documents = DOCUMENTS
    n = len(questions)

    pref = {"q": "", "d": ""}
    for key, v in MODEL_PREFIXES.items():
        if key in model_name:
            pref = v
            break

    # Encodage questions
    t0 = time.perf_counter()
    if "Snowflake" in model_name:
        emb_q = model.encode(
            questions,
            prompt_name="query",
            batch_size=CFG.batch_size,
            prefix_query=pref["q"],
            prefix_doc=pref["d"]
        )
    else:
        emb_q = model.encode(
            [pref["q"] + q for q in questions],
            convert_to_tensor=True,
            batch_size=CFG.batch_size,
            show_progress_bar=True
        )
    t1 = time.perf_counter()

    # Encodage documents
    if "Snowflake" in model_name:
        emb_d = model.encode(
            documents,
            prompt_name=None,
            batch_size=CFG.batch_size,
            prefix_query=pref["q"],
            prefix_doc=pref["d"]
        )
    else:
        emb_d = model.encode(
            [pref["d"] + d for d in documents],
            convert_to_tensor=True,
            batch_size=CFG.batch_size,
            show_progress_bar=True
        )
    t2 = time.perf_counter()

    # Retrieval blockwise
    block_size = CFG.block_size_cuda if DEVICE == "cuda" else CFG.block_size_cpu
    t3 = time.perf_counter()
    topk_idx = compute_topk_blockwise(emb_q, emb_d, k_eval=CFG.k_eval, block_size=block_size)
    t4 = time.perf_counter()

    # Metrics
    metrics = compute_retrieval_metrics_from_topk(
        topk_idx,
        questions,
        mrr_k=CFG.mrr_k,
        ndcg_k=CFG.ndcg_k
    )

    metrics["N"] = n
    metrics["Encode Q time (s)"] = t1 - t0
    metrics["Encode D time (s)"] = t2 - t1
    metrics["Retrieval time (s)"] = t4 - t3
    metrics["Total compute time (s)"] = t4 - t0
    return metrics

# 6) TABLEAU BENCHMARK
def display_ranked_table(df_sorted: pd.DataFrame):
    gradient_cols = [
        "R@1 (%)",
        "R@5 (%)",
        "R@10 (%)",
        "R@20 (%)",
        f"MRR@{CFG.mrr_k}",
        f"nDCG@{CFG.ndcg_k}",
        "TotalCompute(s)",
    ]
    gradient_cols = [c for c in gradient_cols if c in df_sorted.columns]

    df_display = df_sorted[["Modèle"] + gradient_cols].copy().reset_index(drop=True)

    fmt = {
        "R@1 (%)": "{:.1f}",
        "R@5 (%)": "{:.1f}",
        "R@10 (%)": "{:.1f}",
        "R@20 (%)": "{:.1f}",
        f"MRR@{CFG.mrr_k}": "{:.3f}",
        f"nDCG@{CFG.ndcg_k}": "{:.3f}",
        "TotalCompute(s)": "{:.3f}",
    }
    fmt = {k: v for k, v in fmt.items() if k in df_display.columns}

    styled = (
        df_display.style
        .format(fmt)
        .hide(axis="index")
        .set_caption("🏆 Classement par métriques (retrieval)")
        .set_properties(**{
            "background-color": "#1e1e1e",
            "color": "white",
            "border-color": "#333333",
            "text-align": "center",
            "font-size": "12pt"
        })
        .set_table_styles([
            {"selector": "th", "props": [
                ("background-color", "#111111"),
                ("color", "white"),
                ("border-color", "#333333"),
                ("text-align", "center")
            ]},
            {"selector": "td", "props": [
                ("border-color", "#333333")
            ]},
            {"selector": "caption", "props": [
                ("caption-side", "top"),
                ("color", "white"),
                ("font-size", "14pt"),
                ("font-weight", "bold")
            ]}
        ])
    )

    styled = styled.set_properties(
        subset=["Modèle"],
        **{
            "background-color": "#1f3c88",
            "color": "white",
            "font-weight": "bold"
        }
    )
    display(styled)

# 7) EXECUTION
results = []

if not DATASET_QA:
    print("Erreur : Dataset vide.")
else:
    print(f"\nDémarrage du Benchmark (N={len(DATASET_QA)})...")

    for cand in CANDIDATS:
        gc.collect()
        cuda_empty_cache()

        print(f"\nChargement : {cand['name']}...")
        model, load_time, err = load_model_smart(cand)

        if err is not None or model is None:
            print(f"Erreur Load {cand['name']}: {err}")
            continue

        try:
            metrics = run_benchmark(model, cand["name"])
            print(
                f"{cand['name']} -> "
                f"R@1: {metrics['R@1 (%)']:.1f}% | "
                f"R@5: {metrics['R@5 (%)']:.1f}% | "
                f"R@10: {metrics['R@10 (%)']:.1f}% | "
                f"R@20: {metrics['R@20 (%)']:.1f}% | "
                f"MRR@{CFG.mrr_k}: {metrics[f'MRR@{CFG.mrr_k}']:.3f} | "
                f"nDCG@{CFG.ndcg_k}: {metrics[f'nDCG@{CFG.ndcg_k}']:.3f} | "
            )

            row = {
                "Modèle": cand["name"],
                "R@1 (%)": metrics["R@1 (%)"],
                "R@5 (%)": metrics["R@5 (%)"],
                "R@10 (%)": metrics["R@10 (%)"],
                "R@20 (%)": metrics["R@20 (%)"],
                f"MRR@{CFG.mrr_k}": metrics[f"MRR@{CFG.mrr_k}"],
                f"nDCG@{CFG.ndcg_k}": metrics[f"nDCG@{CFG.ndcg_k}"],
                "EncQ(s)": metrics["Encode Q time (s)"],
                "EncD(s)": metrics["Encode D time (s)"],
                "Retr(s)": metrics["Retrieval time (s)"],
                "TotalCompute(s)": metrics["Total compute time (s)"],
            }
            results.append(row)

        except Exception as e:
            print(f"Erreur benchmark {cand['name']}: {e}")

        del model
        gc.collect()
        cuda_empty_cache()

    if not results:
        print("Aucun résultat valide.")
    else:
        df = pd.DataFrame(results)

        df_sorted = df.sort_values(by="R@1 (%)", ascending=False).reset_index(drop=True)
        display_ranked_table(df_sorted)

        metric_cols = ["R@1 (%)", "R@5 (%)", "R@10 (%)", "R@20 (%)"]
        df_melted = df_sorted.melt(id_vars=["Modèle"], value_vars=metric_cols, var_name="Métrique", value_name="Score")

        plt.figure(figsize=(12, 6))
        ax = sns.barplot(data=df_melted, x="Modèle", y="Score", hue="Métrique")
        plt.title(f"Performance Retrieval (Recall) sur {len(DATASET_QA)} paires - Benchmark", fontsize=14)
        plt.ylim(0, 105)
        plt.grid(axis="y", linestyle="--", alpha=0.5)
        plt.xlabel("")
        plt.xticks(rotation=15)
        for container in ax.containers:
            ax.bar_label(container, fmt="%.1f", padding=3)
        plt.tight_layout()
        plt.show()

        time_cols = ["EncQ(s)", "EncD(s)", "Retr(s)", "TotalCompute(s)"]
        df_time = df_sorted[["Modèle"] + time_cols].melt(id_vars=["Modèle"], var_name="Phase", value_name="Temps (s)")

        plt.figure(figsize=(12, 6))
        ax2 = sns.barplot(data=df_time, x="Modèle", y="Temps (s)", hue="Phase")
        plt.title("Temps par phase (Encodage / Retrieval / Total compute)", fontsize=14)
        plt.grid(axis="y", linestyle="--", alpha=0.5)
        plt.xlabel("")
        plt.xticks(rotation=15)
        plt.tight_layout()
        plt.show()
```

```python
import os
import sys
import gc
import time
import random
from dataclasses import dataclass
from typing import Dict, List, Optional

import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

os.environ["WANDB_DISABLED"] = "true"
os.environ["TOKENIZERS_PARALLELISM"] = "false"
os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"

def install(package: str):
    os.system(f"{sys.executable} -m pip install -q {package}")

try:
    import datasets
    from sentence_transformers import SentenceTransformer
    from transformers import AutoModel, AutoTokenizer, AutoConfig
except ImportError:
    print("Installation des dépendances...")
    install("sentence-transformers datasets pandas matplotlib seaborn scikit-learn transformers")
    import datasets
    from sentence_transformers import SentenceTransformer
    from transformers import AutoModel, AutoTokenizer, AutoConfig

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Mode : {'GPU' if DEVICE == 'cuda' else 'CPU'}")

def set_all_seeds(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_all_seeds(42)

def cuda_empty_cache():
    if DEVICE == "cuda":
        torch.cuda.empty_cache()

# CONFIG
@dataclass
class Config:
    max_samples: int = None
    batch_size: int = 32
    max_length: int = 512
    block_size_cuda: int = 128
    block_size_cpu: int = 32
    k_eval: int = 20
    mrr_k: int = 10
    ndcg_k: int = 10

CFG = Config()

# 1) WRAPPER SNOWFLAKE (Transformers natif)
class SnowflakeWrapper:
    """
    Wrapper pour utiliser Snowflake (HF transformers) avec une API encode() proche SentenceTransformer.
    Utilise mean pooling + normalisation L2.
    """
    def __init__(self, model_name: str):
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        config = AutoConfig.from_pretrained(model_name, trust_remote_code=True)

        # Fix anti-crash potentiels (si présents)
        if hasattr(config, "use_memory_efficient_attention"):
            config.use_memory_efficient_attention = False
        if hasattr(config, "unpad_inputs"):
            config.unpad_inputs = False

        self.model = AutoModel.from_pretrained(
            model_name,
            config=config,
            add_pooling_layer=False,
            trust_remote_code=True
        ).to(DEVICE)
        self.model.eval()

    @staticmethod
    def _mean_pooling(last_hidden_state: torch.Tensor, attention_mask: torch.Tensor) -> torch.Tensor:
        mask = attention_mask.unsqueeze(-1).type_as(last_hidden_state)
        summed = (last_hidden_state * mask).sum(dim=1)
        denom = mask.sum(dim=1).clamp(min=1e-9)
        return summed / denom

    def encode(
        self,
        sentences: List[str],
        prompt_name: Optional[str] = None,
        batch_size: int = 32,
        show_progress_bar: bool = False,
        prefix_query: str = "query: ",
        prefix_doc: str = ""
    ) -> torch.Tensor:
        prefix = prefix_query if prompt_name == "query" else prefix_doc
        inputs = [prefix + s for s in sentences]

        all_embeddings = []
        for i in range(0, len(inputs), batch_size):
            batch_texts = inputs[i:i + batch_size]
            batch_tokens = self.tokenizer(
                batch_texts,
                padding=True,
                truncation=True,
                return_tensors="pt",
                max_length=CFG.max_length
            ).to(DEVICE)

            with torch.inference_mode():
                outputs = self.model(**batch_tokens)
                emb = self._mean_pooling(outputs[0], batch_tokens["attention_mask"])
                emb = torch.nn.functional.normalize(emb, p=2, dim=1)
                all_embeddings.append(emb)

        if all_embeddings:
            return torch.cat(all_embeddings, dim=0)
        return torch.empty((0, 0), device=DEVICE)

# 2) DATASET (100% - 1 terme = 1 définition)
print("\nChargement du dataset...")
try:
    dataset = datasets.load_dataset("baconnier/administration_multi-term-dataset", split="train")

    cols = dataset.column_names
    col_terme = next(c for c in cols if "term" in c.lower())
    col_def = next(c for c in cols if "def" in c.lower())

    # 1 TERME -> 1 DEF (définition canonique = la plus fréquente)
    from collections import Counter, defaultdict as dd

    TERM_TO_DEF_COUNTER = dd(Counter)

    for row in dataset:
        terms = row[col_terme]
        defs = row[col_def]

        m = min(len(terms), len(defs))
        for i in range(m):
            t = str(terms[i]).strip().lower()
            d = str(defs[i]).strip()
            if not t or not d:
                continue
            TERM_TO_DEF_COUNTER[t][d] += 1

    TERM_TO_DEF = {}
    for t, counter in TERM_TO_DEF_COUNTER.items():
        best_def, _ = counter.most_common(1)[0]
        TERM_TO_DEF[t] = best_def

    TERMS = sorted(TERM_TO_DEF.keys())
    DOCUMENTS = [TERM_TO_DEF[t] for t in TERMS]

    TERM_TO_INDICES = {t: [i] for i, t in enumerate(TERMS)}

    QUESTIONS = [f"C'est quoi {t} ?" for t in TERMS]
    DATASET_QA = [(q, "") for q in QUESTIONS]

    print(f"Nb questions (termes uniques) : {len(QUESTIONS)}")
    print(f"Nb documents (defs) : {len(DOCUMENTS)}")

    t0 = TERMS[0]
    print("Exemple terme :", t0)
    print("Index def :", TERM_TO_INDICES[t0])
    print("Extrait def :", DOCUMENTS[TERM_TO_INDICES[t0][0]][:160], "...")

except Exception as e:
    print(f"Erreur Dataset : {e}")
    DATASET_QA = []

# 3) CANDIDATS + PREFIXES
CANDIDATS = [
    {"name": "Solon-Large", "id": "OrdalieTech/SOLON-embeddings-large-0.1"},
    {"name": "Snowflake-Arctic", "id": "Snowflake/snowflake-arctic-embed-m-v2.0"},
    {"name": "BGE-M3", "id": "BAAI/bge-m3"},
    {"name": "E5-Large", "id": "intfloat/multilingual-e5-large"},
    {"name": "E5-Large-instruct", "id": "intfloat/multilingual-e5-large-instruct"},
    {"name": "GTE-Multilingual", "id": "Alibaba-NLP/gte-multilingual-base", "trust": True},
    {"name": "E5-Small", "id": "intfloat/multilingual-e5-small"},
]

MODEL_PREFIXES: Dict[str, Dict[str, str]] = {
    "E5": {"q": "query: ", "d": "passage: "},
    "Snowflake": {"q": "query: ", "d": ""},
}

# 4) METRICS + TOPK BLOCKWISE
def compute_retrieval_metrics_from_topk(
    topk_indices: torch.Tensor,
    questions: List[str],
    mrr_k: int = 10,
    ndcg_k: int = 10
) -> Dict[str, float]:
    n, K = topk_indices.shape
    r1 = r5 = r10 = r20 = 0
    mrr = 0.0
    ndcg = 0.0

    k5 = min(5, K)
    k10 = min(10, K)
    k20 = min(20, K)
    mrr_k = min(mrr_k, K)
    ndcg_k = min(ndcg_k, K)

    for i in range(n):
        term = (
            questions[i]
            .replace("C'est quoi", "")
            .replace("?", "")
            .strip()
            .lower()
        )

        relevant = set(TERM_TO_INDICES[term])
        ranked = topk_indices[i].tolist()

        if ranked[0] in relevant:
            r1 += 1
        if any(idx in relevant for idx in ranked[:k5]):
            r5 += 1
        if any(idx in relevant for idx in ranked[:k10]):
            r10 += 1
        if any(idx in relevant for idx in ranked[:k20]):
            r20 += 1

        rr = 0.0
        for rank_pos, doc_idx in enumerate(ranked[:mrr_k], start=1):
            if doc_idx in relevant:
                rr = 1.0 / rank_pos
                break
        mrr += rr

        dcg = 0.0
        for rank_pos, doc_idx in enumerate(ranked[:ndcg_k], start=1):
            if doc_idx in relevant:
                dcg += 1.0 / np.log2(rank_pos + 1)

        rel_count = min(len(relevant), ndcg_k)
        idcg = 0.0
        for rank_pos in range(1, rel_count + 1):
            idcg += 1.0 / np.log2(rank_pos + 1)

        ndcg += (dcg / idcg) if idcg > 0 else 0.0

    return {
        "R@1 (%)": (r1 / n) * 100,
        "R@5 (%)": (r5 / n) * 100,
        "R@10 (%)": (r10 / n) * 100,
        "R@20 (%)": (r20 / n) * 100,
        f"MRR@{mrr_k}": (mrr / n),
        f"nDCG@{ndcg_k}": (ndcg / n),
    }

def compute_topk_blockwise(
    emb_q: torch.Tensor,
    emb_d: torch.Tensor,
    k_eval: int,
    block_size: int
) -> torch.Tensor:
    nq = emb_q.size(0)
    nd = emb_d.size(0)
    K = min(k_eval, nd)
    emb_d_t = emb_d.T

    out_topk = []
    for i in range(0, nq, block_size):
        q_block = emb_q[i:i + block_size]
        scores = torch.matmul(q_block, emb_d_t)
        topk = torch.topk(scores, k=K, dim=1).indices
        out_topk.append(topk.detach().cpu())

        del scores, topk, q_block
        cuda_empty_cache()

    return torch.cat(out_topk, dim=0)

# 5) LOADER + BENCHMARK (SANS CACHE)
def load_model_smart(cand):
    t0 = time.perf_counter()
    model = None
    err = None

    try:
        if "Snowflake" in cand["name"]:
            model = SnowflakeWrapper(cand["id"])
        else:
            kw = {"torch_dtype": torch.float16} if DEVICE == "cuda" else {}
            trust = cand.get("trust", False)
            model = SentenceTransformer(cand["id"], trust_remote_code=trust, device=DEVICE, model_kwargs=kw)
    except Exception as e:
        err = e

    t1 = time.perf_counter()
    load_time = t1 - t0
    return model, load_time, err

def run_benchmark(model, model_name: str) -> Dict[str, float]:
    questions = QUESTIONS
    documents = DOCUMENTS
    n = len(questions)

    pref = {"q": "", "d": ""}
    for key, v in MODEL_PREFIXES.items():
        if key in model_name:
            pref = v
            break

    # Encodage questions
    t0 = time.perf_counter()
    if "Snowflake" in model_name:
        emb_q = model.encode(
            questions,
            prompt_name="query",
            batch_size=CFG.batch_size,
            prefix_query=pref["q"],
            prefix_doc=pref["d"]
        )
    else:
        emb_q = model.encode(
            [pref["q"] + q for q in questions],
            convert_to_tensor=True,
            batch_size=CFG.batch_size,
            show_progress_bar=True
        )
    t1 = time.perf_counter()

    # Encodage documents
    if "Snowflake" in model_name:
        emb_d = model.encode(
            documents,
            prompt_name=None,
            batch_size=CFG.batch_size,
            prefix_query=pref["q"],
            prefix_doc=pref["d"]
        )
    else:
        emb_d = model.encode(
            [pref["d"] + d for d in documents],
            convert_to_tensor=True,
            batch_size=CFG.batch_size,
            show_progress_bar=True
        )
    t2 = time.perf_counter()

    # Retrieval blockwise
    block_size = CFG.block_size_cuda if DEVICE == "cuda" else CFG.block_size_cpu
    t3 = time.perf_counter()
    topk_idx = compute_topk_blockwise(emb_q, emb_d, k_eval=CFG.k_eval, block_size=block_size)
    t4 = time.perf_counter()

    # Metrics
    metrics = compute_retrieval_metrics_from_topk(
        topk_idx,
        questions,
        mrr_k=CFG.mrr_k,
        ndcg_k=CFG.ndcg_k
    )

    metrics["N"] = n
    metrics["Encode Q time (s)"] = t1 - t0
    metrics["Encode D time (s)"] = t2 - t1
    metrics["Retrieval time (s)"] = t4 - t3
    metrics["Total compute time (s)"] = t4 - t0
    return metrics

# 6) TABLEAU BENCHMARK
def display_ranked_table(df_sorted: pd.DataFrame):
    gradient_cols = [
        "R@1 (%)",
        "R@5 (%)",
        "R@10 (%)",
        "R@20 (%)",
        f"MRR@{CFG.mrr_k}",
        f"nDCG@{CFG.ndcg_k}",
        "TotalCompute(s)",
    ]
    gradient_cols = [c for c in gradient_cols if c in df_sorted.columns]

    df_display = df_sorted[["Modèle"] + gradient_cols].copy().reset_index(drop=True)

    fmt = {
        "R@1 (%)": "{:.1f}",
        "R@5 (%)": "{:.1f}",
        "R@10 (%)": "{:.1f}",
        "R@20 (%)": "{:.1f}",
        f"MRR@{CFG.mrr_k}": "{:.3f}",
        f"nDCG@{CFG.ndcg_k}": "{:.3f}",
        "TotalCompute(s)": "{:.3f}",
    }
    fmt = {k: v for k, v in fmt.items() if k in df_display.columns}

    styled = (
        df_display.style
        .format(fmt)
        .hide(axis="index")
        .set_caption("🏆 Classement par métriques (retrieval)")
        .set_properties(**{
            "background-color": "#1e1e1e",
            "color": "white",
            "border-color": "#333333",
            "text-align": "center",
            "font-size": "12pt"
        })
        .set_table_styles([
            {"selector": "th", "props": [
                ("background-color", "#111111"),
                ("color", "white"),
                ("border-color", "#333333"),
                ("text-align", "center")
            ]},
            {"selector": "td", "props": [
                ("border-color", "#333333")
            ]},
            {"selector": "caption", "props": [
                ("caption-side", "top"),
                ("color", "white"),
                ("font-size", "14pt"),
                ("font-weight", "bold")
            ]}
        ])
    )

    styled = styled.set_properties(
        subset=["Modèle"],
        **{
            "background-color": "#1f3c88",
            "color": "white",
            "font-weight": "bold"
        }
    )
    display(styled)

# 7) EXECUTION
results = []

if not DATASET_QA:
    print("Erreur : Dataset vide.")
else:
    print(f"\nDémarrage du Benchmark (N={len(DATASET_QA)})...")

    for cand in CANDIDATS:
        gc.collect()
        cuda_empty_cache()

        print(f"\nChargement : {cand['name']}...")
        model, load_time, err = load_model_smart(cand)

        if err is not None or model is None:
            print(f"Erreur Load {cand['name']}: {err}")
            continue

        try:
            metrics = run_benchmark(model, cand["name"])
            print(
                f"{cand['name']} -> "
                f"R@1: {metrics['R@1 (%)']:.1f}% | "
                f"R@5: {metrics['R@5 (%)']:.1f}% | "
                f"R@10: {metrics['R@10 (%)']:.1f}% | "
                f"R@20: {metrics['R@20 (%)']:.1f}% | "
                f"MRR@{CFG.mrr_k}: {metrics[f'MRR@{CFG.mrr_k}']:.3f} | "
                f"nDCG@{CFG.ndcg_k}: {metrics[f'nDCG@{CFG.ndcg_k}']:.3f} | "
            )

            row = {
                "Modèle": cand["name"],
                "R@1 (%)": metrics["R@1 (%)"],
                "R@5 (%)": metrics["R@5 (%)"],
                "R@10 (%)": metrics["R@10 (%)"],
                "R@20 (%)": metrics["R@20 (%)"],
                f"MRR@{CFG.mrr_k}": metrics[f"MRR@{CFG.mrr_k}"],
                f"nDCG@{CFG.ndcg_k}": metrics[f"nDCG@{CFG.ndcg_k}"],
                "EncQ(s)": metrics["Encode Q time (s)"],
                "EncD(s)": metrics["Encode D time (s)"],
                "Retr(s)": metrics["Retrieval time (s)"],
                "TotalCompute(s)": metrics["Total compute time (s)"],
            }
            results.append(row)

        except Exception as e:
            print(f"Erreur benchmark {cand['name']}: {e}")

        del model
        gc.collect()
        cuda_empty_cache()

    if not results:
        print("Aucun résultat valide.")
    else:
        df = pd.DataFrame(results)

        df_sorted = df.sort_values(by="R@1 (%)", ascending=False).reset_index(drop=True)
        display_ranked_table(df_sorted)

        metric_cols = ["R@1 (%)", "R@5 (%)", "R@10 (%)", "R@20 (%)"]
        df_melted = df_sorted.melt(id_vars=["Modèle"], value_vars=metric_cols, var_name="Métrique", value_name="Score")

        plt.figure(figsize=(12, 6))
        ax = sns.barplot(data=df_melted, x="Modèle", y="Score", hue="Métrique")
        plt.title(f"Performance Retrieval (Recall) sur {len(DATASET_QA)} paires - Benchmark", fontsize=14)
        plt.ylim(0, 105)
        plt.grid(axis="y", linestyle="--", alpha=0.5)
        plt.xlabel("")
        plt.xticks(rotation=15)
        for container in ax.containers:
            ax.bar_label(container, fmt="%.1f", padding=3)
        plt.tight_layout()
        plt.show()

        time_cols = ["EncQ(s)", "EncD(s)", "Retr(s)", "TotalCompute(s)"]
        df_time = df_sorted[["Modèle"] + time_cols].melt(id_vars=["Modèle"], var_name="Phase", value_name="Temps (s)")

        plt.figure(figsize=(12, 6))
        ax2 = sns.barplot(data=df_time, x="Modèle", y="Temps (s)", hue="Phase")
        plt.title("Temps par phase (Encodage / Retrieval / Total compute)", fontsize=14)
        plt.grid(axis="y", linestyle="--", alpha=0.5)
        plt.xlabel("")
        plt.xticks(rotation=15)
        plt.tight_layout()
        plt.show()
```
